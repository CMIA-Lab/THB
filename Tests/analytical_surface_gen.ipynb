```python
import numpy as np
import matplotlib.pyplot as plt
```

## xysinxcosy

```python
N = 5
x = np.outer(np.linspace(-N, N, 100), np.ones(100))
y = x.copy().T
z = x*y*np.sin(x)*np.cos(y)

xyz = np.vstack([x.flatten(), y.flatten(), z.flatten()]).T
print(xyz.shape)
xyz = (xyz-xyz.min())/(xyz.max()-xyz.min())
np.savetxt('surfaces/xysincos.xyz', xyz, delimiter=' ')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
plt.show()
```

## Gaussian Bell

```python
def gaussian(x, y):
    return N*np.exp(-x**2)*np.exp(-y**2)

N = 4
x = np.outer(np.linspace(-N, N, 100), np.ones(100))
y = x.copy().T
z = gaussian(x, y)

xyz = np.vstack([x.flatten(), y.flatten(), z.flatten()]).T
xyz = (xyz-xyz.min())/(xyz.max()-xyz.min())
print(xyz.shape)

np.savetxt('bell/target100x100.xyz', xyz, delimiter=' ')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
plt.show()
```

## Three Peaks example from 2012_Gianelli

```python
import numpy as np
import matplotlib.pyplot as plt

def three_peaks(x, y):
    return 2/(3*np.exp(((10*x-3)**2+(10*y-3)**2)**0.5)) + 2/(3*np.exp(((10*x+3)**2+(10*y+3)**2)**0.5)) + 2/(3*np.exp((10*x**2+10*y**2)**0.5))

N = 1
x = np.outer(np.linspace(-N, N, 512), np.ones(512))
y = x.copy().T
z = three_peaks(x, y)

xyz = np.vstack([x.flatten(), y.flatten(), z.flatten()]).T
xyz = (xyz-xyz.min())/(xyz.max()-xyz.min())
print(xyz.shape)

np.savetxt('peaks/target512x512.xyz', xyz, delimiter=' ')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z, s=0.01)
plt.show()
```

### Surface Generation for Multipatch Fitting

```python
N = 5
x = np.outer(np.linspace(0, N, 100), np.ones(100))
y = np.outer(np.ones(100), np.linspace(0, -N, 100))
z = x*y*np.sin(x)*np.cos(y)

xyz = np.vstack([x.flatten(), y.flatten(), z.flatten()]).T
print(xyz.shape)
xyz = (xyz-xyz.min())/(xyz.max()-xyz.min())
np.savetxt('multipatch/xysincos_q4.xyz', xyz, delimiter=' ')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
plt.show()
```
